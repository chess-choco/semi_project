```python
import chess.svg
```

```python
print('''캐슬 (Castle, O-O / O-O-O)

캐슬은 킹을 룩 뒤에 숨기는 방법으로, 게임 중 1번만 할 수 있습니다.
킹을 좌측이나 우측으로 2칸 움직이고, 룩을 킹이 온 방향 쪽 옆구리에 붙이면 됩니다.

킹사이드 캐슬 (O-O):''')
board = chess.Board("8/8/8/8/8/8/8/4K2R w - - 0 1")
chess.svg.board(
    board,
    arrows=[chess.svg.Arrow(chess.E1, chess.G1, color="#0000cccc")],
    size=200)
```

```python
print('''캐슬을 할 수 있는 조건은 다음과 같습니다.

1. 킹과 룩이 게임이 시작하고 단 한 번도 자리에서 움직인 적이 없어야 합니다.
백 기준 왼쪽(퀸사이드)으로 캐슬을 하려고 하면, 킹과 퀸사이드 룩이 움직인 적이 없어야 합니다.

2. 킹과 룩 사이에 다른 기물이 없어야 합니다. 아군과 적군 기물 모두 포함입니다.

3. 킹이 움직일 2칸이 상대 기물의 공격 범위에 들어가 있지 않아야 하고,
킹이 공격받고 있지 않는 상태여야 합니다. 룩은 공격받고 있어도 캐슬에 참여할 수 있습니다.''')
```

```python
print('''승급, 프로모션 (Promotion, =Q/N/B/R)

폰이 전진하여 자신 기준 8번째 줄, 즉 상대 진영 첫 줄까지 도달하면,
해당 폰을 퀸, 나이트, 비숍, 룩 중 원하는 것으로 승격할 수 있게 됩니다.''')

board = chess.Board("8/4P3/8/8/8/8/3p4/8 w - - 0 1")
chess.svg.board(
    board,
    arrows=[chess.svg.Arrow(chess.D2, chess.D1, color="#0000cccc"),
    chess.svg.Arrow(chess.E7, chess.E8, color="#0000cccc")],
    size=200)
```

```python
board = chess.Board("4Q3/8/8/8/8/8/8/3r4 w - - 0 1")
chess.svg.board(
    board,
    size=200)
```

```python
print('''모든 폰을 살려서 퀸으로 승급시킨다면 퀸 9개도 가능합니다!
보통은 가장 강력한 퀸으로 승급시키기 때문에 퀴닝(Queening)이라고 부르기도 합니다.''')
```

```python
print('''앙 파상 (En Passant)

상대 폰이 한 번에 2칸을 전진하여 자신 폰 옆에 도달했다면,
상대방 폰이 1칸만 움직인 것처럼 자신의 폰으로 잡을 수 있는 규칙을 앙파상이라고 부릅니다.
앙파상은 상대방 폰이 2칸을 전진한 그 순간에만 가능하며, 반드시 해야 하는 것은 아닙니다.''')

board = chess.Board("8/8/8/3pP3/8/8/8/8 b - - 0 1")
chess.svg.board(
    board,
    arrows=[chess.svg.Arrow(chess.D7, chess.D5, color="#0000cccc")],
    size=200)
```
